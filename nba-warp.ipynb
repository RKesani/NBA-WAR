```python
# Create a full name column
# And create a column with number of games each player has played
df['playFuNm'] = df['playFNm'] + ' ' + df['playLNm']
df['games_played'] = df.groupby(['playFuNm'])['playFNm'].transform('count')

```

```python
### In order to connect player box scores to team box scores i created a unique game id
# It's the date of the game and the team 

df['teams'] = df.apply(lambda row: [row['teamAbbr'], row['opptAbbr']], axis=1)
dfg['teams'] = dfg.apply(lambda row: [row['teamAbbr'], row['opptAbbr']], axis=1)

df['game_id'] = df['gmDate'].astype(str) + '_' + df.apply(lambda row: '_'.join(sorted(row['teams'])), axis=1)
df['game_id'] = df['game_id'].apply(hash)

dfg['game_id'] = dfg['gmDate'].astype(str) + '_' + dfg.apply(lambda row: '_'.join(sorted(row['teams'])), axis=1)
dfg['game_id'] = dfg['game_id'].apply(hash)


```

```python
### Adding in number of Team Posessions per game

dfg['teamPoss']  = dfg['teamFGA'] - dfg['teamORB'] + dfg['teamTO'] + (0.44 * dfg['teamFTA'])
dfg['opptPoss']  = dfg['opptFGA'] - dfg['opptORB'] + dfg['opptTO'] + (0.44 * dfg['opptFTA'])
```

```python
df['playPoss']  = df['playFGA'] - df['playORB'] + df['playTO'] + (0.44 * df['playFTA'])
df['playPoss'] = df['playPoss'].apply(lambda x: max(0.001, x))

dfg = dfg.drop_duplicates(subset=['game_id', 'teamAbbr'], keep='first')

df_merged = pd.merge(df, dfg, on=['game_id', 'teamAbbr'], how='left', suffixes=('_df', '_dfg'))


```

```python
dff = df_merged
dff['playUsage'] = (dff['playPoss']/dff['playMin']) / ((dff['teamFGA'] + 0.44 * dff['teamFTA'] + dff['teamTO']) / (dff['teamMin'] / 5))
dff.head(30)
```

```python
selected_cols = ['playLNm', 'playPoss', 'teamPoss', 'teamFGA', 'teamFTA', 'teamTO', 'teamMin']
dff[selected_cols].head(30)
```

```python
LgORGmaj = 102.3
LgORGmin = 91.3
unique_ORBp = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamOREB%']]
LgORBp = unique_ORBp['teamOREB%'].mean()
LgORBp = LgORBp / 100
unique_DRBp = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamDREB%']]
LgDRBp = unique_DRBp['teamDREB%'].mean()
LgDRBp = LgDRBp / 100
unique_TO = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamTO']]
LgTO = unique_TO['teamTO'].mean()
unique_Poss = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamPoss']]
LgPoss = unique_Poss['teamPoss'].mean()
unique_FTA = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFTA']]
LgFTA = unique_FTA['teamFTA'].mean()
unique_FTA = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFTA']]
LgFTA = unique_FTA['teamFTA'].mean()
unique_EFG = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamEFG%']]
LgEFG = unique_EFG['teamEFG%'].mean()
unique_FG = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFG%']]
LgFG = unique_FG['teamFG%'].mean()
unique_Reb = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamTRB']]
LgReb = unique_Reb['teamTRB'].mean()
unique_FGA = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFGA']]
LgFGA = unique_FGA['teamFGA'].mean()
unique_FGM = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFGM']]
LgFGM = unique_FGM['teamFGM'].mean()
unique_FTM = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFTM']]
LgFTM = unique_FTM['teamFTM'].mean()
unique_ORB = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamORB']]
LgORB = unique_ORB['teamORB'].mean()
unique_BLK = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamBLK']]
LgBLK = unique_BLK['teamBLK'].mean()
unique_STL = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamSTL']]
LgSTL = unique_STL['teamSTL'].mean()
unique_PF = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamPF']]
LgPF = unique_PF['teamPF'].mean()
unique_TO = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamTO']]
LgTO = unique_TO['teamTO'].mean()
unique_FTM = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamFTM']]
LgFTM = unique_FTM['teamFTM'].mean()
unique_PTS = dff.drop_duplicates(['game_id', 'teamAbbr'])[['game_id', 'teamAbbr', 'teamPTS']]
LgPTS = unique_PTS['teamPTS'].mean()
LgFTAR = LgFTA/LgPF


```

```python
dff['PC'] = dff['playPTS'] + (0.75 * dff['playAST']) - 0.75 * dff['playFGM'] * ((dff['teamAST']/dff['teamFGM']) * (1.53 - 1.442 * ((dff['playAST']/dff['playMin'])/ (dff['teamAST'] / (dff['teamMin'] / 5))) - 0.041 * ((dff['playORB'] / dff['playMin']) * 48) - 0.787 * dff['playUsage'] + 0.014 * ((dff['play3PA'] / dff['playFGA']) ** 2) * ((1/dff['playUsage']) ** 2) ))
dff['playPos'] = dff['playFGA'] + (0.44 * dff['playFTA']) + dff['playTO'] + 0.375 * dff['playAST'] - 0.375 * dff['playFGM'] * ((dff['teamAST']/dff['teamFGM']) * (1.53 - 1.442 * ((dff['playAST']/dff['playMin'])/ (dff['teamAST'] / (dff['teamMin'] / 5))) - 0.041 * ((dff['playORB'] / dff['playMin']) * 48) - 0.787 * dff['playUsage'] + 0.014 * ((dff['play3PA'] / dff['playFGA']) ** 2) * ((1/dff['playUsage']) ** 2) ))
dff['playPos'] = dff['playPos'].apply(lambda x: max(0, x))
dff['playOR'] = 100 * dff['PC']/dff['playPos']
```

```python
selected_cols = ['playLNm', 'playPoss', 'playMin', 'playPTS', 'teamFTA', 'teamTO', 'teamMin']
```

```python
dff['ORmin'] = dff['playOR'] * (dff['playPoss'] / dff['teamPoss']) + ((1 - dff['playPoss'] / dff['teamPoss']) * LgORGmin + (25 * (dff['playPoss'] / dff['teamPoss'] - 0.2)))
df['ORmin'] = dff['ORmin'].fillna(0)
dff.head(30)
dff['ORBp'] = 100 * (dff['playORB'] * (dff['teamMin'] / 5))/(dff['playMin'] * (dff['teamORB'] + dff['opptDRB']))
dff['ORBp'] = dff['ORBp'].fillna(0)
dff['tmORBp'] = (dff['ORBp'] + (4 * LgORBp) / (5 - LgORBp) * (1 - dff['ORBp']))/100
dff['playRebOpp'] = (dff['playFGA'] - dff['playFGM']) + 0.56 * (dff['playFTA'] - dff['playFTM'])/(dff['playFGA'] + 0.44 * dff['playFTA'] + dff['playTO'])
dff['playRebOpp'] = dff['playRebOpp'].fillna(0)
dff['tmOR'] = LgORGmin + (25 * (dff['playPoss'] / dff['teamPoss'] - 0.2))
dff['tmFGAp'] = (1 - ((LgTO / LgPoss) *  (1 + (0.5 * (0.2 - dff['playUsage'])))) - (LgFTA/LgPoss))
dff['tmEFG'] = ((dff['tmOR'] / 100) - (LgFTM / LgPoss))/(1/dff['tmFGAp']) - 0.025
dff['tmFGp'] = dff['tmEFG'] / (LgEFG / LgFG)
dff['tmRebOpp'] = (1 - dff['tmFGp'])*(dff['tmFGAp']) + 0.56 * ((LgFTA - LgFTM)/LgPoss)
dff['tmRebOpp'] = dff['tmRebOpp'].fillna(0)
dff['teamRebOpp'] = (dff['playRebOpp'] * dff['playUsage'] + dff['tmRebOpp'] * (1 - dff['playUsage'])) / 100
RebFactor1 = LgReb / ((LgFGA - LgFGM) + 0.56 * (LgFTA - LgFTM))
RebFactor2 = (LgORGmaj - LgORGmin) / (LgORGmin * LgORB/LgPoss)

dff['Offence'] = ((dff['ORmin'] + (dff['tmORBp'] * dff['teamRebOpp'] * RebFactor1 * RebFactor2 * dff['ORmin']))) + 7 * ((dff['play3PA']/(dff['playFGA'] + 0.44*dff['playFTA'] + dff['playTO'])) - (dff['team3PA']/(dff['teamFGA'] + 0.44*dff['teamFTA'] + dff['teamTO'])))
dff['Offence'] = dff['Offence'].fillna(0)
print(RebFactor1)
print(RebFactor2)

```

```python
selected_cols = ['playLNm', 'ORmin', 'tmORBp', 'teamRebOpp', 'ORBp', 'playTRB', 'Offence', 'playOR', 'playPos']
```

```python
average_scores = dff.groupby(['playLNm', 'playFNm'])['Offence'].mean()
top_10_players = average_scores.nlargest(30)

print(f"The top 10 players with the highest average in are:")
print(top_10_players)

dff[dff['playFNm'] == 'LeBron'][selected_cols].head(40)


```

```python
dff['playDPoss'] = dff['playMin'] * ((dff['opptFGA'] + (0.44 * dff['opptFTA']) + dff['opptTO'])/(dff['teamMin'] / 5))
dff['tmBSPoss'] = (dff['playBLK'] + dff['playSTL']) / dff['playDPoss'] + 0.8 * ((LgBLK + LgSTL)/LgPoss)
dff['tmFTAPoss'] = LgFTAR * (dff['playPF']/dff['playDPoss']) + 0.8 * ((0.44 * LgFTA)/LgPoss)
dff['TDF'] = dff['playMin'] / dff['teamMin']
dff['tmFTOPoss'] = dff['TDF'] * ((dff['opptTO'] - dff['teamSTL'])/dff['opptPoss']) + (1 - dff['TDF']) * ((LgTO - LgSTL) / LgPoss)
dff['tmFGRat'] = (dff['TDF'] * ((dff['opptPTS'] - dff['opptFTM'])/(dff['opptFGA'] - dff['teamBLK'])) + (1 - dff['TDF']) * ((LgPTS - LgFTM)/(LgFGA - LgBLK)))
dff['tmDRatmin'] = (100 * ((dff['tmFTAPoss'] * ((LgFTM)/(LgFTA * 0.44))) + (1 - dff['tmFTAPoss'] - dff['tmBSPoss'] - dff['tmFTOPoss']) * dff['tmFGRat']))
dff['playDRBp'] = (100 * (dff['playDRB'] * (dff['teamMin'] / 5))/(dff['playMin'] * (dff['teamDRB'] + dff['opptORB'])))/100
dff['tmDRBp'] = (dff['playDRBp'] + 0.8 * LgDRBp * (1 - dff['playDRBp']))/100
dff['opptFGp'] = ((dff['tmFGRat'] / 2) / (LgEFG/LgFG))/100
dff['playBLKp'] = (100 * (dff['playBLK'] * (dff['teamMin'] / 5))/(dff['playMin'] * (dff['opptFGA'] + dff['oppt3PA'])))/100
dff['opptRebOpp'] = ((1 - dff['opptFGp']) * (1 - dff['tmFTAPoss'] - dff['tmBSPoss'] - dff['tmFTOPoss']) + 0.56 * ((LgFTA - LgFTM)/LgPoss) + dff['teamBLK%'] + dff['playBLK']/dff['playDPoss'] + 0.8 * LgBLK/LgPoss)/100
dff['Defence'] = dff['tmDRatmin'] + (1 - dff['tmDRBp']) * dff['opptRebOpp'] * RebFactor1 * RebFactor2*dff['tmDRatmin']
dff['Defence'] = dff['Defence'].fillna(35)

selected_cols = ['playLNm', 'playFNm', 'playMin', 'playBLK', 'playSTL', 'playDPoss', 'tmBSPoss', 'tmFTAPoss', 'TDF', 'tmFTOPoss', 'tmFGRat', 'tmDRatmin', 'playDRBp', 'tmDRBp', 'opptFGp', 'playBLKp', 'opptRebOpp', 'Defence']



```

```python
average_scores = dff.groupby(['playLNm', 'playFNm'])['Defence'].mean()
top_10_players = average_scores.nsmallest(40)


print(f"The top 10 players with the highest average in are:")
print(top_10_players)

dff[dff['playLNm'] == 'Curry']['Defence'].mean()

```
